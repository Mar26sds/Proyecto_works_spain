```python
def extraerlink2(url):
    
    driver.get(url)
    tabla1=driver.find_element(By.XPATH,'//*[@id="main"]/div/div/div[1]')
    a=tabla1.find_elements(By.TAG_NAME,'a')
    classname=tabla1.find_elements(By.CLASS_NAME,'app-aware-link')#[1].get_attribute('href')
    links_to_sql=[]

    for i in range(len(classname)):
        if 'company' in classname[i].get_attribute('href'):
            links_to_sql.append(classname[i].get_attribute('href'))
        else: pass
    links_unique=list(dict.fromkeys(links_to_sql))
    return links_unique
```

```python
#Este sí funciona. Hacer en casa porque se va el wifi


links_final2=[]
posible = True
for url in urls_data:
    print (url)
    while(posible):
        try:    
            links_final2.append(extraerlink2(url)) #función definida
            time.sleep(3)
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)
            url1=driver.current_url
            driver.find_element(By.XPATH,'//button[@type="button" and @aria-label="Siguiente"]').click()
            time.sleep(3)
            url=driver.current_url
            
            if url==url1:
                break
        except: posible=False



print('ha parado el bucle')
#el resultado es links_final
```

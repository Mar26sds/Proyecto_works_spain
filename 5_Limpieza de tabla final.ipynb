```python
data.drop(['Link'], axis=1, inplace=True)
```

```python
#Para la columna 'Lugar' voy a poner las ciudades en mayúsculas y bien escritas
```

```python
data.Lugar.value_counts()
```

```python
for i in range(len(data.Lugar)):
    data.loc[i,'Lugar'] = data.loc[i,'Lugar'].title()
    if data.Lugar[i]=='Madrid':
        data.loc[i,'Lugar'] = 'Comunidad de Madrid'
    elif data.Lugar[i]=='Valenciana':
        data.loc[i,'Lugar'] = 'Comunidad Valenciana'
    elif data.Lugar[i]=='Paisvasco':
        data.loc[i,'Lugar'] = 'País Vasco'
    elif data.Lugar[i]=='Castillaleon':
        data.loc[i,'Lugar'] = 'Castilla y León'
    elif data.Lugar[i]=='Aragon':
        data.loc[i,'Lugar'] = 'Aragón'
    elif data.Lugar[i]=='Castillamancha':
        data.loc[i,'Lugar'] = 'Castilla la Mancha'
    elif data.Lugar[i]=='Larioja':
        data.loc[i,'Lugar'] = 'La Rioja'
```

```python
data.Lugar.value_counts()
```

```python
#Columna 'especialidad' hay diferentes datos, la especialidad, unknown y la fecha que tiene que ir en fundación. Vamos a solucionarlo.
```

```python
def if_integer(string):
    try: 
        int(string)
        return True
    except ValueError:
        return False
```

```python
fundacion_añadir=[]
especialidad_=[]

for i in range(len(data.especialidad)):
    try: 
        int(data.especialidad[i])
        fundacion_añadir.append(data.especialidad[i])
        especialidad_.append('Unknown')
    except ValueError:
        fundacion_añadir.append('Unknown')
        especialidad_.append(data.especialidad[i])
        
```

```python
fundacion_añadir
```

```python
especialidad_
```

```python
#Arreglar la columna sector, ya que hay valores que no deberían estar.
```

```python
sector_añadir=[]
tamaño_tenemos=[]

for i in range(len(data.sector)):
    if data.sector[i].endswith('empleados'):
        sector_añadir.append('Unknown')
        tamaño_tenemos.append(data.sector[i])
    else:
        sector_añadir.append(data.sector[i])
        tamaño_tenemos.append('Unknown')
```

```python
tamaño_tenemos
```

```python
# Arreglar la columna 'tamaño-empresa'
```

```python
tamaño_añadir=[]

for i in range(len(data.tamaño_empresa)):
    if 'en' in data.tamaño_empresa[i]:
        tamaño_añadir.append('Unknown')
    else:
        tamaño_añadir.append(data.tamaño_empresa[i])
```

```python
tamaño_añadir
```

```python
# Arreglar la columna 'fundacion'
```

```python
fundacion_tenemos=[]

for i in range(len(data.fundacion)):
    try: 
        int(data.fundacion[i])
        fundacion_tenemos.append(data.fundacion[i])
    except ValueError:
        fundacion_tenemos.append('Unknown')
        
```

```python
fundacion_tenemos
```

```python
#Crear las columnas completas de fundacion, sector, tamaño empresa
```
